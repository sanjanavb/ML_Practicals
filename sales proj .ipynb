```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Sales Data
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr'],
    'Sales': [12000, 15000, 17000, 13000],
    'Profit': [3000, 4500, 5000, 3500]
}
df = pd.DataFrame(data)

# Plotting Monthly Sales and Profit Trend
plt.figure(figsize=(10, 6))
sns.lineplot(x='Month', y='Sales', data=df, marker='o', label='Sales')
sns.lineplot(x='Month', y='Profit', data=df, marker='o', label='Profit')
plt.title('Monthly Sales and Profit Trend')
plt.legend()
plt.grid(True)
plt.show()

# Regional Sales Data
region_data = {
    'Region': ['North', 'South', 'East', 'West'],
    'Sales': [50000, 60000, 45000, 70000]
}
region_df = pd.DataFrame(region_data)

# Plotting Regional Sales
plt.figure(figsize=(8, 5))
sns.barplot(x='Region', y='Sales', data=region_df, palette='viridis')
plt.title('Regional Sales Performance')
plt.show()

# Category Sales Data
category_data = {
    'Category': ['Electronics', 'Furniture', 'Clothing'],
    'Sales': [30000, 20000, 50000]
}
category_df = pd.DataFrame(category_data)

# Plotting Sales by Category
plt.figure(figsize=(8, 5))
sns.barplot(x='Category', y='Sales', data=category_df, palette='coolwarm')
plt.title('Sales by Category')
plt.show()
```
